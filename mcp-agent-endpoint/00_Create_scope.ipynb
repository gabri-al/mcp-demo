```python
import os
from dotenv import load_dotenv
load_dotenv()  # reads variables from a .env file and sets them in os.environ`
print(os.getenv("SCOPE_NAME"))
```

```python
from databricks.sdk import WorkspaceClient
w = WorkspaceClient()

# List all scopes in the workspace
scopes = w.secrets.list_scopes()
scope_names = [n.name for n in scopes]

# Check if scope exists
# w.secrets.delete_scope(scope=os.getenv("SCOPE_NAME")) # Pre-delete scope if required

proceed = True if os.getenv("SCOPE_NAME") not in scope_names else False
print(proceed)
```

```python
# Put information into the scope
if proceed:

  # Create scope
  w.secrets.create_scope(scope=os.getenv("SCOPE_NAME"))

  # Put info into the scope
  w.secrets.put_secret(
    scope=os.getenv("SCOPE_NAME"),
    key="DATABRICKS_CLIENT_ID",
    string_value=os.getenv("DATABRICKS_CLIENT_ID")
  )

  w.secrets.put_secret(
    scope=os.getenv("SCOPE_NAME"),
    key="DATABRICKS_CLIENT_SECRET",
    string_value=os.getenv("DATABRICKS_CLIENT_SECRET")
  )

else:
  print("Information already saved into the scope")
```
